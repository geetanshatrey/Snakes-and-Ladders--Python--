```python
# Snakes and Ladders Game
# Author : Geetansh Atrey
#

import random

snakes_position = {
    17: 7,
    54: 34,
    62: 19,
    98: 79
}

ladders_position = {
    3: 38,
    24: 33,
    42: 93,
    72: 84
} 


def welcome_msg():
    msg = """
    #### Welcome to Snakes & Ladders Game ####
    Developed by: Geetansh Atrey

    Rules:
      1. Initially both the players start at the position 0.
      2. You have two choices : Auto Mode or Manual Mode.
      3. For Auto Mode, enter "roll". For Manual Mode, enter a number between 1 and 20.
      2. If you land up at the bottom of a ladder, you climb up to its top.
      3. If you land up on the head of a snake, you slide down to its bottom.
      4. The first player to get to the final position : 100, is the winner.
      5. Enter "exit" to exit the game at any instant.

    """
    print(msg)
    

def get_player_names():
    player1_name = None
    while player1_name == None:
        player1_name = input('Enter the name of Player 1: ')

    player2_name = None
    while player2_name == None:
        player2_name = input('Enter the name of Player 2: ')
    
    return player1_name, player2_name


def check_ladder_jump(player_position):
    if player_position in ladders_position.keys():
            if ladders_position[player_position] >= 0:
                player_position = ladders_position[player_position]
    return player_position


def check_snake_bite(player_position):
    if player_position in snakes_position.keys():
            if snakes_position[player_position] >= 0:
                player_position = snakes_position[player_position]
    return player_position
            
    
def auto_mode(players, player_name, player_position, winner):
    rolledDice = random.randrange(1, 7)
    print("You got a " + str(rolledDice))
    
    if player_position + rolledDice < 100:
        player_position += rolledDice
        player_position = check_ladder_jump(player_position)
        player_position = check_snake_bite(player_position)
    
    elif player_position + rolledDice == 100:
        player_position += rolledDice
        winner = players[player_name]
    
    return player_position, winner
    

def manual_mode(players, player_name, player_position, input_from_player, winner):
    input_number = int(input_from_player)
    print("You got a ",input_number)
    
    if input_number > 1 and input_number < 20:
        
        if player_position + input_number < 100:
            player_position += input_number
            player_position = check_ladder_jump(player_position)
            player_position = check_snake_bite(player_position)
            
        elif player_position + input_number == 100:
            player_position += input_number
            winner = players[player_name]
    
    return player_position, winner


def check_game_over(input_from_player, winner):
    return input_from_player != "quit" and winner == -1


def game_start():
    welcome_msg()
    player1_name, player2_name = get_player_names()
    
    players = {
        player1_name: 1,
        player2_name: 2,
    }
    
    player_one_position = 0
    player_two_position = 0

    input_of_player_one = ""
    input_of_player_two = ""

    winner = -1

    print('#### Let us Start ####')
    
    while True:
        input_of_player_one = input("Player 1: ")
        
        if input_of_player_one == "roll":
            player_one_position, winner = auto_mode(players, player1_name, player_one_position, winner)
            
        elif input_of_player_one.isnumeric():
            player_one_position, winner = manual_mode(players, player1_name, player_one_position, input_of_player_one, winner)
            
        print("Your final position is ", player_one_position)
        if check_game_over(input_of_player_one, winner) == False:
            break
        
        
        input_of_player_two = input("Player 2: ")
        
        if input_of_player_two == "roll":
            player_two_position, winner = auto_mode(players, player2_name, player_two_position, winner)
            
        elif input_of_player_two.isnumeric():
            player_two_position, winner = manual_mode(players, player2_name, player_two_position, input_of_player_two, winner)
            
        print("Your final position is ", player_two_position)
        if check_game_over(input_of_player_two, winner) == False:
            break
    
    
    print("Player " + str(winner) + " won the game")
    print('#### Game Successfully Finished ####')
    
    

if __name__ == "__main__":
    game_start()        
```
